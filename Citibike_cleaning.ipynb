```python
import os
import pandas as pd
```

```python
main_df = pd.DataFrame()

# loop through all CSV files
for subdir, dirs, files in os.walk("2020-citibike-tripdata"):
    for file in files:
        filepath = os.path.join(subdir, file)
        
        # check if CSV data file
        if (".csv" in filepath):
            # make sample df
            df = pd.read_csv(filepath, low_memory=False)
            df = df.sample(frac=0.02, random_state=42)
            main_df = pd.concat([main_df, df], axis=0)
            print(filepath)
    
main_df = main_df.reset_index(drop=True)
```

```python
main_df.columns
```

```python
main_df.head()
```

```python
main_df.info()
```

```python
main_df["started_at"] = pd.to_datetime(main_df.started_at)
main_df["ended_at"] = pd.to_datetime(main_df.ended_at)

main_df.head()
```

```python
main_df.info()
```

```python
main_df.to_csv("2020_citibike_sample.csv", index=False)
```
